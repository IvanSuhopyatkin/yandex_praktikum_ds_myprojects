## Предобработка данных

переведем тип first_day_exposition в datetime:

```python
data['first_day_exposition']=pd.to_datetime(data['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')
```

и проверим что данные изменились:

```python
print(data['first_day_exposition'].head(5))
display(data.info())
```

заменим в is_apartment NaN на False:


```python
data['is_apartment']=data['is_apartment'].fillna(False)
```

и проверим что все успешно:

```python
print(data['is_apartment'].isna().sum()) 
```

```python
#и аналогично с balcony, но заменяем на 0:
data['balcony']=data['balcony'].fillna(0)
#и проверим что все успешно:
print(data['balcony'].isna().sum()) 
```

```python
#и также с days_exposition:
print(data['days_exposition'].isna().sum()) 
data['days_exposition']=data['days_exposition'].fillna(0)
#и проверим что все успешно:
print(data['days_exposition'].isna().sum()) 
```

остальные пропуски заменять не будем, так как нам точно неизвестна их причина 

```python
#изменим тип данных у переменных:
#last_price переведем в int потому что десятые доли нам не интересны
data['last_price']=data['last_price'].astype('int')
#аналогично и с числом балконов:
data['balcony']=data['balcony'].astype('int')
#аналогично и с числом балконов:
data['parks_around3000']=data['parks_around3000'].astype('Int64')
#аналогично и с числом аэропортов:
data['airports_nearest']=data['airports_nearest'].astype('Int64') 
#аналогично с parks_nearest, доли метра до ближайшего парка нам не нужно:
data['parks_nearest']=data['parks_nearest'].astype('Int64') 
#число водоемов не бывает нецелым:
data['ponds_around3000']= data['ponds_around3000'].astype('Int64')
#доли метра до ближайшего парка нам не нужно:
data['ponds_nearest']= data['ponds_nearest'].astype('Int64')
#доли дней от публикации до снятия нам не нужны:
data['days_exposition']=data['days_exposition'].astype('int')
```

тип floors_total и cityCenters_nearest не стал менять потому что если поменять на int то возникает ошибка, а если на Int64 то все NaN меняются на NA и тогда перестает работать функция what_floor. 

проверим что все данные поменялись:

```python
display(data.head(20))
```

## Расчёты и добавление результатов в таблицу

```python
#цена квадратного метра это последняя цена деленная на общую площадь квартиры
data['price_per_square_meter']=data['last_price']/data['total_area']
```

```python
#добавим день недели:
data['day']=pd.DatetimeIndex(data['first_day_exposition']).weekday
#добавим месяц:
data['month']=pd.DatetimeIndex(data['first_day_exposition']).month
#добавим год:
data['year']=pd.DatetimeIndex(data['first_day_exposition']).year
```

```python
#напишем функцию которая определяет этаж квартиры: первый, последний, другой;

def what_floor(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'

data['what_floor'] = data.apply(what_floor, axis = 1)
```

```python
#соотношение жилой и общей площади, а также отношение площади кухни к общей.

#добавим соотношение жилой площади к общей
data['ratio_living_total'] = data['living_area'] / data['total_area']

#добавим столбец отношения площади кухни к общей
data['ratio_kitchen_total'] = data['kitchen_area'] / data['total_area']
```

```python
#проверим что все данные успешно добавились в таблицу
display(data.head(10))
```

для простоты будем считать цену квартир в тысячах:

```python
data['last_price']=data['last_price']/1000
```
